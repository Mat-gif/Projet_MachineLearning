```python
import pandas as pd
from pandas import *
data_allemand = pd.read_csv('FakeNews_Task3_2022\FakeNews_Task3_2022\Task3_Test\German_data_test_release_with_rating.csv', sep=',')
data_allemand_other = data_allemand.loc[~data_allemand['our rating'].isin(['mixture', 'Partially False', 'False', 'True'])] 
display(data_allemand_other['our rating'].value_counts())

v = data_allemand_other['text'].str.len()
v = v.to_frame()
print(v['text'].sum())

z = data_allemand_other['title'].str.len()
z = z.to_frame()
print(z['title'].sum())
```

```python
import deepl

auth_key = "clé d'authentification après avoir crée un compte gratuit pour l'api je ne donne pas mon accès veuillez en créer une"
translator = deepl.Translator(auth_key)()

result = translator
tab_title = []

for title in data_allemand_other["title"] : 
    result = translator.translate_text(title,source_lang="de",target_lang="EN-US")
    tab_title.append(result.text)
    data_allemand_other = data_allemand_other.replace(to_replace=title, value=result.text)
    
    
display(data_allemand_other)
```
